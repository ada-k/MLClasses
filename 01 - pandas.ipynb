### Series and DataFrames
* A series is basically a column. Can be created from lists, tuples, ...
* A dataframe is an nD series. Can be created from lists, tuples, numpy arrays, pandas series, dictionaries...

```python
[1,2,3]
(1,2,3)
```

```python
# series
a =pd.Series([1,2,3]) 
b = pd.Series((4,5,6))
print(a)
print(b)
```

```python
# dataframes

# from series
c = pd.DataFrame((a, b))
# c
# from lists
d = pd.DataFrame([['ada', 21, 'Kenya'], ['Ola', 23, "USA"]])
# d
# from dictionary
data = {
    'apples': [3, 2, 0, 1], 
    'oranges': [0, 3, 7, 2]}
e = pd.DataFrame(data, index = ['abubakar', 'abdualaziz', 'nahom', 'nabil'])
# e
# from tuples
f = pd.DataFrame((('ada', 21, 'Kenya'), ('Ola', 23, "USA")))
# f
print('from pandas series\n', c, '\n\n from a list of lists\n', d, '\n\n from python dict\n', e, '\n\n from tuple of tuples\n', f)
```

### Pandas Indexing


* **Dataframe.[[ ]]** ; indexing operator
* **Dataframe.loc[ ]** : used for labels.
* **Dataframe.iloc[ ]** : used for positions or integer based
* **Dataframe.ix[]** : used for both label and integer based


```python
g = e[["apples"]] 
h = e.loc['abubakar']
i = e.iloc[:,1]
j = e.ix["nahom"] # depreceated
print(g, '\n\n', h, '\n\n', i, '\n\n', j)
```
